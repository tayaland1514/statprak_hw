## Problem 2


Write a function that takes a two-dimensional array and a string as input and returns an array rotated 90 degrees counterclockwise if the string 'left' was passed, and clockwise if the string 'right' was passed.

Example for input: $\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$.\
If the string 'left' is passed, the function should return $\begin{bmatrix} 3 & 6 & 9 \\ 2 & 5 & 8 \\ 1 & 4 & 7 \end{bmatrix}$, and if the string 'right' is passed, the function should return $\begin{bmatrix} 7 & 4 & 1 \\ 8 & 5 & 2 \\ 9 & 6 & 3 \end{bmatrix}$.

Do not use numpy and nested loops. Use list comprehensions and built-in function `zip`.

````python
print("Введите количество строк матрицы:")
rows = int(input())
matrix = []
print("Введите матрицу:")
for i in range(rows):
    matrix.append(list(map(int, input().split())))
print("\nВведите left или right:")
direction = input()

if direction == 'right':
    res = [list(row)[::-1] for row in zip(*matrix)]
elif direction == 'left':
    res = [list(row) for row in list(zip(*matrix))[::-1]]
for row in res:
    print(row)
````

## Problem 3

Write a function that takes a string as input and returns a dictionary containing the number of occurrences of each character in the string. Use method `count` of strings. Implement this function in one line using $\lambda$-function.

## Problem 4

### Implementing a Library Management System

#### Description

You are required to design and implement a system for managing books and users in a library. The system should allow for the management of books (adding, deleting, searching by various criteria) and users (registration, deletion, searching), as well as tracking the history of interactions between them (issuing and returning books).

#### Tasks

1. **`Book` Class**:
   - Attributes: title, author, year of publication, ISBN, number of copies.
   - Methods: constructor, methods to get information about the book, method to change the number of copies (when issuing and returning books).

2. **`User` Class**:
   - Attributes: user name, library card number, list of borrowed books.
   - Methods: constructor, methods for user registration, methods for adding and removing books from the borrowed list.

3. **`Library` Class**:
   - Attributes: list of books, list of users, transaction history (who, when, which book was borrowed and returned).
   - Methods: constructor, methods for adding and deleting books and users, methods for issuing and returning books, searching for books and users by various criteria, method to display the transaction history.

#### Assignment

1. Implement the `Book`, `User`, and `Library` classes with the specified attributes and methods.
2. Create several books and users, and add them to the library system.
3. Implement scenarios for issuing books to users and their return.
4. Display the transaction history to show how books were issued and returned.


````python

class Book:
    def __init__(self, title, author, year, isbn, copies):
        self.title = title
        self.author = author
        self.year = year
        self.isbn = isbn
        self.copies = copies
        print(f"Book '{title}' created")

    def get_info(self):
        return f"Title: '{self.title}', author: {self.author}, year: {self.year}, ISBN: {self.isbn}, copies: {self.copies}"

    def change_copies(self, delta):
        self.copies += delta
        print(f"Copies of '{self.title}' changed; now: {self.copies}")
        return True


class User:
    def __init__(self, name, card_number):
        self.name = name
        self.card_number = card_number
        self.borrowed_books = []
        print(f"User {name} created")

    def register(self):
        return f"User {self.name} (card: {self.card_number}) registered"

    def add_book(self, book):
        if book not in self.borrowed_books and book.copies > 0:
            self.borrowed_books.append(book)
            print(f"{self.name} borrowed '{book.title}'")
            return True
        print(f"{self.name} cannot borrow '{book.title}'")
        return False

    def remove_book(self, book):
        if book in self.borrowed_books:
            self.borrowed_books.remove(book)
            print(f"{self.name} returned '{book.title}'")
            return True
        print(f"Return failed. {self.name} doesn't have '{book.title}'")
        return False

    def list_borrowed_books(self):
        return [book.get_info() for book in self.borrowed_books]

    def get_info(self):
        return f"{self.name} (card: {self.card_number}) - Books borrowed: {len(self.borrowed_books)}"


class Library:
    def __init__(self, name):
        self.name = name
        self.books = []
        self.users = []
        self.transaction_history = []
        print(f"Library '{name}' created")

    def add_book(self, book):
        self.books.append(book)
        self.transaction_history.append(f"Book added: {book.get_info()}")

    def remove_book(self, book):
        if book in self.books and book.copies == 0:
            self.books.remove(book)
            self.transaction_history.append(f"Book removed: {book.get_info()}")
            return True
        return False

    def register_user(self, user):
        self.users.append(user)
        self.transaction_history.append(f"User registered: {user.register()}")

    def deregister_user(self, user):
        if user in self.users and len(user.borrowed_books) == 0:
            self.users.remove(user)
            self.transaction_history.append(f"User removed: {user.name} (card: {user.card_number})")
            return True
        return False

    def issue_book(self, user, book):
        if user not in self.users or book not in self.books:
            return False

        if user.add_book(book) and book.change_copies(-1):
            self.transaction_history.append(f"Book issued: '{book.title}' to {user.name} (card: {user.card_number})")
            return True
        return False

    def return_book(self, user, book):
        if user.remove_book(book) and book.change_copies(1):
            self.transaction_history.append(
                f"Book returned: '{book.title}' from {user.name} (card: {user.card_number})")
            return True
        return False

    def find_books(self, **criteria):
        results = []
        for book in self.books:
            match = True
            for key, value in criteria.items():
                if key == 'title' and value.lower() not in book.title.lower():
                    match = False
                elif key == 'author' and value.lower() not in book.author.lower():
                    match = False
                elif key == 'year' and book.year != value:
                    match = False
                elif key == 'isbn' and book.isbn != value:
                    match = False
            if match:
                results.append(book)
        return results

    def find_users(self, **criteria):
        results = []
        for user in self.users:
            match = True
            for key, value in criteria.items():
                if key == 'name' and value.lower() not in user.name.lower():
                    match = False
                elif key == 'card_number' and user.card_number != value:
                    match = False
            if match:
                results.append(user)
        return results

    def display_info(self):
        return f"Library '{self.name}' - Books: {len(self.books)}, Users: {len(self.users)}"

    def display_transaction_history(self):
        print(f"\nTransaction History for '{self.name}':")
        if not self.transaction_history:
            print("  No transactions yet")
            return
        for i, record in enumerate(self.transaction_history, 1):
            print(f"  {i}. {record}")

library = Library("My Library")
print()

book1 = Book("Gravity Falls", "Alex Hirsch", 2016, "978-1-23456-789-0", 3)
library.add_book(book1)
book2 = Book("Taras Pavlovich Lukashenko and philosopher's stone", "Philosopher's stone", 2021, "978-1-23456-788-3", 2)
library.add_book(book2)
book3 = Book("Machine Learning", "Machine Learner", 2022, "978-1-23456-787-6", 1)
library.add_book(book3)

assert len(library.books) == 3, "Library should have 3 books"
assert book1 in library.books, "Book1 should be in library"
assert book2 in library.books, "Book2 should be in library"
assert book3 in library.books, "Book3 should be in library"
print()

user1 = User("Dzhigan", "CARD-001")
library.register_user(user1)
user2 = User("Oksana Samoylova", "CARD-002")
library.register_user(user2)
user3 = User("Ilia Malinin", "CARD-003")
library.register_user(user3)

assert len(library.users) == 3, "Library should have 3 users"
assert user1 in library.users, "User1 should be in library"
assert user2 in library.users, "User2 should be in library"
assert user3 in library.users, "User3 should be in library"
print()

print("Testing book issuing:")
assert library.issue_book(user1, book1) == True, "User1 should be able to borrow book1"
assert len(user1.borrowed_books) == 1, "User1 should have 1 book"
assert book1.copies == 2, "Book1 copies should decrease to 2"

assert library.issue_book(user1, book2) == True, "User1 should be able to borrow book2"
assert len(user1.borrowed_books) == 2, "User1 should have 2 books"
assert book2.copies == 1, "Book2 copies should decrease to 1"

assert library.issue_book(user2, book1) == True, "User2 should be able to borrow book1"
assert len(user2.borrowed_books) == 1, "User2 should have 1 book"
assert book1.copies == 1, "Book1 copies should decrease to 1"

# Try to issue the same book to same user
assert library.issue_book(user1, book1) == False, "User1 should not borrow the same book twice"

# Try to issue last copy of book3
assert library.issue_book(user3, book3) == True, "User3 should borrow book3"
assert book3.copies == 0, "Book3 copies should be 0"

# Try to issue book with no copies
assert library.issue_book(user2, book3) == False, "Should not borrow book with 0 copies"
print()

print("Testing book returning:")
assert library.return_book(user1, book1) == True, "User1 should return book1"
assert len(user1.borrowed_books) == 1, "User1 should have 1 book left"
assert book1.copies == 2, "Book1 copies should increase to 2"

assert library.return_book(user1, book2) == True, "User1 should return book2"
assert len(user1.borrowed_books) == 0, "User1 should have 0 books"
assert book2.copies == 2, "Book2 copies should increase to 2"

# Try to return book user doesn't have (should fail)
assert library.return_book(user1, book3) == False, "Should not return book user doesn't have"

# Return book1 from user2
assert library.return_book(user2, book1) == True, "User2 should return book1"
assert len(user2.borrowed_books) == 0, "User2 should have 0 books"
assert book1.copies == 3, "Book1 copies should be back to 3"

# Return book3 from user3
assert library.return_book(user3, book3) == True, "User3 should return book3"
assert len(user3.borrowed_books) == 0, "User3 should have 0 books"
assert book3.copies == 1, "Book3 copies should be back to 1"
print()

print("Testing searching:")
# Find books by author
found_books = library.find_books(author="Hirsch")
assert len(found_books) == 1, "Should find 1 book by Hirsch"
assert found_books[0].title == "Gravity Falls", "Should find Gravity Falls"

# Find books by title
found_books = library.find_books(title="Taras")
assert len(found_books) == 1, "Should find 1 book with 'Taras' in title"
assert found_books[0].title == "Taras Pavlovich Lukashenko and philosopher's stone", "Should find Taras Pavlovich Lukashenko and philosopher's stone"

# Find users by name
found_users = library.find_users(name="Ilia")
assert len(found_users) == 1, "Should find 1 user named Ilia"
assert found_users[0].card_number == "CARD-003", "Should find Ilia with correct card"

# Find users by card number
found_users = library.find_users(card_number="CARD-002")
assert len(found_users) == 1, "Should find 1 user with card CARD-002"
assert found_users[0].name == "Oksana Samoylova", "Should find Oksana Samoylova"
print("Search tests passed")
print()

print("Testing book removal:")
# Make book3 have 0 copies by issuing it again
library.issue_book(user3, book3)
# Now book3 has 0 copies, try to remove it
assert library.remove_book(book3) == True, "Should remove book with 0 copies"
assert book3 not in library.books, "Book3 should not be in library"
print()

print("Testing user deregistration:")
# Try to deregister user with books (should fail - user3 has book3)
assert library.deregister_user(user3) == False, "Should not deregister user with books"

# Return book3 first
library.return_book(user3, book3)

# Now try to deregister user with no books
assert library.deregister_user(user2) == True, "Should deregister user with no books"
assert user2 not in library.users, "User2 should not be in library"

# Try to deregister user not in library
assert library.deregister_user(user2) == False, "Should not deregister user not in library"
print()

print("Testing display methods:")
print(library.display_info())

print("\nAll books in library:")
for book in library.books:
    print(f"{book.get_info()}")

print("\nAll registered users:")
for user in library.users:
    print(f"{user.get_info()}")

library.display_transaction_history()
print()


assert len(library.books) == 2, "Library should have 2 books (book3 was removed)"
assert len(library.users) == 2, "Library should have 2 users (user2 was deregistered)"
assert book1.copies == 3, "Book1 should have 3 copies"
assert book2.copies == 2, "Book2 should have 2 copies"
assert book3.copies == 1, "Book3 should have 1 copy (returned before removal test)"
assert len(user1.borrowed_books) == 0, "User1 should have 0 books"
assert len(user3.borrowed_books) == 0, "User3 should have 0 books"
````

## Problem 6*

Let
$$A = \sum_{i=1}^{10000} \frac{1}{i^2},\quad B=\sum_{i=10000}^{1} \frac{1}{i^2}.$$
Calculate the values of $A$ and $B$ and compare them. What do you observe? Explain why this happens. What is the best way to calculate the value of $\sum\limits_{i=1}^{10000} \dfrac{1}{i^2}$?

````python

n = 10000
A = 0.0
B = 0.0
for i in range(1, n + 1):
    A += 1.0 / (i * i)
for i in range(n, 0, -1):
    B += 1.0 / (i * i)

print(A)
print(B)
print(f"{A - B:.16f}")
print()
print("Если складывать от 1 до 10000:")
print('Память для хранения числа (в нашем случае с плавающей точкой) ограничена конкретным кол-вом бит, и не все цифры после запятой влезут в эту ячейку, так что точность этого "хвоста" из цифр после запятой теряется при добавлении маленьких чисел к большим (сначала в сумме идут большие числа, потом маленькие)')
print()
print("Если складывать от 10000 до 1:")
print('В начале суммы будут маленькие числа, они будут складываться между собой, перед тем, как добавиться к большой сумме, и результат будет точнее => этот способ лучше')
````
