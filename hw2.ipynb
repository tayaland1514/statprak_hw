## Problem 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def beauty_derivative_plot(x_point):
    """
    Строим график функции f(x) = 2sin(x) - cos(2x), её производной
    и касательной в точке x_point
    """
    def f(x):
        return 2 * np.sin(x) - np.cos(2 * x)
    def f_derivative(x):
        return 2 * np.cos(x) + 2 * np.sin(2 * x)

    x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
    y = f(x)
    y_derivative = f_derivative(x)
    y_point = f(x_point)
    slope_point = f_derivative(x_point)

    tangent_line = y_point + slope_point * (x - x_point) # ур-е касательной

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8)) # два подграфика друг под другом

    ax1.plot(x, y, 'b-', linewidth=2, label='f(x) = 2sin(x) - cos(2x)')
    ax1.plot(x, tangent_line, 'r--', linewidth=2, label=f'Касательная в точке x = {x_point:.2f}')
    ax1.plot(x_point, y_point, 'ro', markersize=8, label=f'Точка касания ({x_point:.2f}, {y_point:.2f})')
    ax1.set_xlabel('x', fontsize=12)
    ax1.set_ylabel('y', fontsize=12)
    ax1.set_title('Функция и касательная', fontsize=14, fontweight='bold')
    ax1.legend(fontsize=10)
    ax1.grid(True, alpha=0.3)
    ax1.axhline(y=0, color='k', linewidth=0.5, alpha=0.5)
    ax1.axvline(x=0, color='k', linewidth=0.5, alpha=0.5)

    ax2.plot(x, y_derivative, 'g-', linewidth=2, label="f'(x) = 2cos(x) + 2sin(2x)")
    ax2.plot(x_point, slope_point, 'ro', markersize=8, label=f'Значение производной в точке: {slope_point:.2f}')
    ax2.set_xlabel('x', fontsize=12)
    ax2.set_ylabel("f'(x)", fontsize=12)
    ax2.set_title('Производная', fontsize=14, fontweight='bold')
    ax2.legend(fontsize=10)
    ax2.grid(True, alpha=0.3)
    ax2.axhline(y=0, color='k', linewidth=0.5, alpha=0.5)
    ax2.axvline(x=0, color='k', linewidth=0.5, alpha=0.5)

    plt.tight_layout()
    plt.show()

    print(f"\nДля точки x = {x_point:.4f}")
    print(f"f(x) = {y_point:.4f}")
    print(f"f'(x) = {slope_point:.4f}")
```

```python
beauty_derivative_plot(0)
```

```python
beauty_derivative_plot(1)
```

```python
beauty_derivative_plot(np.pi / 2)
```

## Problem 2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('populations.txt')

years = data[:, 0].astype(int)
hares = data[:, 1]
lynxes = data[:, 2]
carrots = data[:, 3]
print(f"   Годы: {years[0]} - {years[-1]}")
print(f"   Всего записей: {len(years)}")
```

```python
# График 1 - динамика численности всех видов
plt.figure(figsize=(12, 6))
plt.plot(years, hares, 'o-', linewidth=2, markersize=6, label='Зайцы')
plt.plot(years, lynxes, 's-', linewidth=2, markersize=6, label='Рыси')
plt.plot(years, carrots, '^-', linewidth=2, markersize=6, label='Морковь')
plt.xlabel('Год', fontsize=12)
plt.ylabel('Численность', fontsize=12)
plt.title('Динамика популяций в Северной Канаде (1900-1920)', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
max_hare_year = years[np.argmax(hares)]
max_lynx_year = years[np.argmax(lynxes)]
max_carrot_year = years[np.argmax(carrots)]
print("Годы максимальной популяции:")
print(f"Зайцы: {max_hare_year} (численность: {np.max(hares):.0f})")
print(f"Рыси: {max_lynx_year} (численность: {np.max(lynxes):.0f})")
print(f"Морковь: {max_carrot_year} (численность: {np.max(carrots):.0f})")
```

```python
# График 2 -динамика с отмеченными максимумами
plt.figure(figsize=(12, 6))
plt.plot(years, hares, 'o-', label='Зайцы', color='blue', alpha=0.7)
plt.plot(years, lynxes, 's-', label='Рыси', color='orange', alpha=0.7)
plt.plot(years, carrots, '^-', label='Морковь', color='green', alpha=0.7)
plt.plot(max_hare_year, np.max(hares), 'ro', markersize=12, label=f'Макс. зайцев ({max_hare_year})', markeredgecolor='darkred')
plt.plot(max_lynx_year, np.max(lynxes), 'rs', markersize=12, label=f'Макс. рысей ({max_lynx_year})', markeredgecolor='darkred')
plt.plot(max_carrot_year, np.max(carrots), 'r^', markersize=12, label=f'Макс. моркови ({max_carrot_year})', markeredgecolor='darkred')
plt.xlabel('Год', fontsize=12)
plt.ylabel('Численность', fontsize=12)
plt.title('Динамика популяций с отмеченными максимумами', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# График 3 - столбчатая диаграмма
total_hares_lynxes = hares + lynxes
hares_pct = (hares / total_hares_lynxes) * 100
lynxes_pct = (lynxes / total_hares_lynxes) * 100
plt.figure(figsize=(14, 6))
bar_width = 0.8
bars1 = plt.bar(years, hares_pct, bar_width, label='Зайцы %', color='green', alpha=0.7)
bars2 = plt.bar(years, lynxes_pct, bar_width, bottom=hares_pct, label='Рыси %', color='orange', alpha=0.7)
plt.xlabel('Год', fontsize=12)
plt.ylabel('Процент от общей численности', fontsize=12)
plt.title('Соотношение популяций зайцев и рысей по годам', fontsize=14, fontweight='bold')
plt.xticks(years, rotation=45)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# статистический анализ
species_names = ['Зайцы', 'Рыси', 'Морковь']
species_data = [hares, lynxes, carrots]
means = [np.mean(s) for s in species_data]
stds = [np.std(s, ddof=1) for s in species_data]
for i, name in enumerate(species_names):
    print(f"{name}:")
    print(f"Среднее = {means[i]:.0f}")
    print(f"Стандартное отклонение = {stds[i]:.0f}")
    print()
```

```python
# График 4 - гистограммы распределения
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for i, ax in enumerate(axes):
    ax.hist(species_data[i], bins=8, alpha=0.7, color='skyblue', edgecolor='black', linewidth=1.2)
    ax.axvline(means[i], color='red', linestyle='dashed', linewidth=2, label=f'Среднее: {means[i]:.0f}')
    ax.set_title(species_names[i], fontsize=12, fontweight='bold')
    ax.set_xlabel('Численность', fontsize=10)
    ax.set_ylabel('Частота (по годам)', fontsize=10)
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.suptitle('Распределение численности популяций', fontsize=14, fontweight='bold', y=1.05)
plt.show()
```

```python
dominant_species = np.where(hares > lynxes, 'H', 'L')
result_string = ''.join(dominant_species)
print(f"Результат: {result_string}")
h_count = np.sum(dominant_species == 'H')
l_count = np.sum(dominant_species == 'L')
print(f"\nЗайцы доминировали: {h_count} лет")
print(f"Рыси доминировали: {l_count} лет")
```

```python
# График 5 - круговая диаграмма доминирования
plt.figure(figsize=(6, 6))
colors = ['#ff9999', '#66b3ff']
plt.pie([h_count, l_count],
        labels=[f'Зайцы (H)\n{h_count} лет', f'Рыси (L)\n{l_count} лет'],
        autopct='%1.1f%%',
        colors=['blue', 'orange'],
        startangle=90,
        textprops={'fontsize': 12})
plt.title('Соотношение лет с доминированием', fontsize=14, fontweight='bold', pad=20)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# График 6 - изменение популяций
hare_change = np.diff(hares)
lynx_change = np.diff(lynxes)
change_years = years[1:]

plt.figure(figsize=(12, 5))
plt.plot(change_years, hare_change, 'o-', linewidth=2, markersize=6, label='Изменение численности зайцев', color='blue')
plt.plot(change_years, lynx_change, 's-', linewidth=2, markersize=6, label='Изменение численности рысей', color='orange')
plt.axhline(y=0, color='gray', linestyle='--', linewidth=1, alpha=0.7)
plt.xlabel('Год', fontsize=12)
plt.ylabel('Изменение численности', fontsize=12)
plt.title('Ежегодное изменение популяций', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

correlation = np.corrcoef(hares, lynxes)[0, 1]
print(f"Коэффициент корреляции между численностью зайцев и рысей: {correlation:.3f}")

```

```python
# График 7 - диаграмма рассеивания
plt.figure(figsize=(10, 8))
scatter = plt.scatter(hares, lynxes, c=years, cmap='viridis', s=150, alpha=0.7, edgecolor='black', linewidth=1)
for i, year in enumerate(years):
    plt.annotate(str(year), (hares[i], lynxes[i]), fontsize=9, fontweight='bold', xytext=(5, 5), textcoords='offset points')

plt.colorbar(scatter, label='Год')
plt.xlabel('Численность зайцев', fontsize=12)
plt.ylabel('Численность рысей', fontsize=12)
plt.title('Корреляция популяций зайцев и рысей', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
print("Оценка численности зайцев в 1920 году")

# среднее за последние 2 года
estimated_hare_1920_mean_last2 = np.mean(hares[-3:-1])
print(f"Оценка через среднее за 1918-1919: {estimated_hare_1920_mean_last2:.0f}")

# взвешенное среднее
weights = [0.1, 0.2, 0.3, 0.4]  # веса для 1916,1917,1918,1919 (чем ближе год, тем он важнее)
weighted_mean = np.average(hares[-5:-1], weights=weights)
print(f"Оценка через взвешенное среднее за 1918-1919: {weighted_mean:.0f}")

true_value = hares[-1]
print(f"\nРеальное значение в 1920 году: {true_value:.0f}")

error_mean2 = np.abs(estimated_hare_1920_mean_last2 - true_value)
error_weighted = np.abs(weighted_mean - true_value)

print(f"Метод 1: ошибка = {error_mean2:.0f} ({error_mean2/true_value*100:.1f}%)")
print(f"Метод 2: ошибка = {error_weighted:.0f} ({error_weighted/true_value*100:.1f}%)")
```
